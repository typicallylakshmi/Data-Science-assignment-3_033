1.Write a Python program to draw line charts of the financial data of Alphabet Inc. between October 3, 2016 to October 7, 2016. Sample Financial data (fdata.csv): Date,Open,High,Low,Close 10-03-16,774.25,776.065002,769.5,772.559998 10-04-16,776.030029,778.710022,772.890015,776.429993 10-05-16,779.309998,782.070007,775.650024,776.469971 10-06-16,779,780.47998,775.539978,776.859985 10-07-16,779.659973,779.659973,770.75,775.080017

```python
import pandas as pd
import matplotlib.pyplot as plt
csv_data = """Date,Open,High,Low,Close
10-03-16,774.25,776.065002,769.5,772.559998
10-04-16,776.030029,778.710022,772.890015,776.429993
10-05-16,779.309998,782.070007,775.650024,776.469971
10-06-16,779,780.47998,775.539978,776.859985
10-07-16,779.659973,779.659973,770.75,775.080017
"""
with open('fdata.csv', 'w') as f:
    f.write(csv_data)
data = pd.read_csv('fdata.csv')
data['Date'] = pd.to_datetime(data['Date'], format='%m-%d-%y')
data.set_index('Date', inplace=True)
plt.figure()
plt.plot(data.index, data['Open'], label='Open')
plt.plot(data.index, data['High'], label='High')
plt.plot(data.index, data['Low'], label='Low')
plt.plot(data.index, data['Close'], label='Close')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Alphabet Inc. Financial Data (Oct 3–7, 2016)')
plt.legend()
plt.show()
```

Q6. Visit data.gov.in , search for the following in “catalogs” option of the website: • Final population Totals, India and states • State Wise literacy rate Download them and create a CSV file containing population data and literacy rate of the respective state. Also add a column Region to the CSV file that should contain the values East, West, North and South. Plot a scatter plot for each region where X axis should be population and Y axis should be Literacy rate. Change the marker to a diamond and size as the square root of the literacy rate. Group the data on the column region and display a bar chart depicting average literacy rate for each region.

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
data = {
    'State': ['Uttar Pradesh','Maharashtra','Bihar','Tamil Nadu','Karnataka','West Bengal','Gujarat','Punjab'],
    'Population': [199812341,112374333,104099452,72147030,61095297,91276115,60439692,27743338],
    'Literacy': [67.68,82.34,61.80,80.09,75.36,76.26,78.03,75.84],
    'Region': ['North','West','East','South','South','East','West','North']
}
df = pd.DataFrame(data)
for region in df['Region'].unique():
    subset = df[df['Region'] == region]

    plt.scatter(
        subset['Population'],
        subset['Literacy'],
        marker='D',
        s=np.sqrt(subset['Literacy']) * 10,
        label=region
    )
plt.xlabel('Population')
plt.ylabel('Literacy Rate')
plt.title('Population vs Literacy Rate by Region')
plt.legend()
plt.show()
avg_lit = df.groupby('Region')['Literacy'].mean()
avg_lit.plot(kind='bar')
plt.title('Average Literacy Rate by Region')
plt.xlabel('Region')
plt.ylabel('Literacy Rate')
plt.show()
```

Q8.Write a Python program to create bar plot of scores by group and gender. Use multiple X values on the same chart for men and women. Sample Data: Means (men) = (22, 30, 35, 35, 26) Means (women) = (25, 32, 30, 35, 29) The code snippet gives the output shown in the following screenshot:

image.png

```python
import numpy as np
import matplotlib.pyplot as plt
men_means = (22, 30, 35, 35, 26)
women_means = (25, 32, 30, 35, 29)
index = np.arange(len(men_means))
bar_width = 0.35
plt.bar(index, men_means, bar_width, label='Men')
plt.bar(index + bar_width, women_means, bar_width, label='Women')
plt.xlabel('Person')
plt.ylabel('Scores')
plt.title('Scores by person')
plt.xticks(index + bar_width / 2, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
products = ["A","B","C","D"]
sales = [150, 200, 300, 250]

# Pie chart
plt.figure()
plt.pie(sales, labels=products, autopct='%1.1f%%')
plt.title("Sales Distribution")
plt.show()

# Subplot
months = ["Jan","Feb","Mar","Apr"]
soap = [50,60,55,70]
facewash = [30,40,35,45]

plt.figure()

plt.subplot(2,1,1)
plt.plot(months, soap)
plt.title("Soap Sales")

plt.subplot(2,1,2)
plt.plot(months, facewash)
plt.title("Facewash Sales")

plt.tight_layout()
plt.show()
```
